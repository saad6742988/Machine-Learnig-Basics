```python
import numpy as np



x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y,learning_rate = 0.05,iterations = 500):
    m_curr = b_curr = 0
    n = len(x)
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))
```

```python
gradient_descent(x,y,0.001,1000)
```

In above output, we start with cost 89 and after 1000 iterations we come to 0.478 cost

### Let's see how changing learning rate and iterations effects

```python
gradient_descent(x,y,0.01,10)
```

0.01 learning rate is also reducing cost, so increase learning rate to 0.1

```python
gradient_descent(x,y,0.1,10)
```

0.1 learning rate cause increase in our cost so we need it between 0.1 and 0.01

```python
gradient_descent(x,y,0.05,10) #Cost Decreased
```

```python
gradient_descent(x,y,0.09,10) #Cost increased
```

```python
gradient_descent(x,y,0.08,10) #Cost Decreased
```

```python
gradient_descent(x,y,0.08,10000) #Cost Decreased
```

So, after a lot of iterations we got some fix and optimal m abd b values with very minimal cost.

## Graphical presentation of changing cost

```python
gradient_descent_Graph(x,y,0.08,35)
```

As, you can see our lines comes near to the actual red points after some iterations.
